```python
import random
```

```python
NUM_QUEENS=8
POPULATION_SAIZE=100
MUTATION_RATE=0.1
NO_GENERATIONS=100
MAX_FITNESS=NUM_QUEENS*(NUM_QUEENS-1)/2
```

```python
def individual():
    return [random.randint(0,NUM_QUEENS-1) for _ in range(NUM_QUEENS)]
```

```python
def fitness(board):
    attacks=0
    for i in range(NUM_QUEENS):
        for j in range (i+1,NUM_QUEENS):
            if board[i]==board[j] or abs(i-j)==abs(board[i]-board[j]):
                attacks+=1
    return NUM_QUEENS*(NUM_QUEENS-1)/2-attacks  #max possible non-attacking pairs        
```

```python
def selection (population):
    selected=[]
    fitness_values = [fitness(individual) for individual in population]
    total_fitness = sum(fitness_values)
    normalized_fitnesses = [f/total_fitness for f in fitness_values]

    for _ in range (POPULATION_SAIZE):
        spin_value = random.random()
        probability_sum = 0
        for i in range(len(population)):
            probability_sum += normalized_fitnesses[i]
            if probability_sum >= spin_value:
                selected.append(population[i])
                break
    return selected        
```

```python
def crossover(parent1,parent2):
    crossover_point = random.randint(1,NUM_QUEENS-1)
    child1 = parent1[:crossover_point] + parent2[crossover_point:]
    child2 = parent2[:crossover_point] + parent1[crossover_point:]
    return child1 , child2
```

```python
def mutation(individual):
    if random.random() < MUTATION_RATE:
        gene_to_mutate = random.randint(0,NUM_QUEENS-1)
        new_value = random.randint(0,NUM_QUEENS-1)
        while new_value == individual[gene_to_mutate]:
            new_value = random.randint(0,NUM_QUEENS-1)
        individual[gene_to_mutate] = new_value
    return individual        
```

```python
def genetic_algorithm():
    population = [individual() for _ in range (POPULATION_SAIZE)]
    best_fitness = 0
    best_individual = None
    for generation in range(NO_GENERATIONS):
        #selection
        selected = selection(population)

        #crossover
        childern = []
        for i in range(0,POPULATION_SAIZE,2):
            child1,child2 =crossover(selected[i],selected[i+1])
            childern.append(child1)
            childern.append(child2)

        #mutation
        for child in childern:
            mutation(child)

        population = childern

        #update best solution
        crrent_fitness = max(fitness(individual) for individual in population) 
        if crrent_fitness > best_fitness:
            best_fitness = crrent_fitness
            best_individual = max(population,key=fitness)
        #print progress (optional)
        print(f"Generation : {generation+1},Best Fitness: {best_fitness}")
        if crrent_fitness == MAX_FITNESS:
            break

    #print sloution 
    print(f"\nSoultion Found in {generation+1} generations:")
    print(f"Board: {best_individual}")
    print(f"Fitness: {best_fitness}") # should be 28 (no queens attacking)    

```

```python
if __name__ == "__main__":
    genetic_algorithm()
```
